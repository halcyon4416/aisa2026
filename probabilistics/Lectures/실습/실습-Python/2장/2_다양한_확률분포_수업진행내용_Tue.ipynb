# 2. 확률분포
## 2.1 확률분포 개념 복습
* 확률분포(probability distribution): 확률변수 $X$가 가질 수 있는 값과 해당하는 확률에 대해 나타낸 것

$$P(a < x < b) = \int^b_af(x)dx$$

* 기대값(expected value):

$$\mu = \text{E}(X) = \int^\infty_{-\infty} x f(x) dx$$

* 분산(varianve)과 표준편차(standard deviation):

$$\text{var}(X) = \text{E}[(X - \mu)^2] = \text{E}(X^2) - (\text{E}(X))^2$$
$$\text{sd}(X) = \sqrt{\text{var}(X)}$$




1.  이항분포: $X \sim \text{Bin}(n, p)$
    * $X_1,X_2,\dots,X_n$은 서로 독립이고, 각각 성공률이 $p$인 베르누이 확률변수이다.
    * $X=X_1+X_2+\dots,+X_n$은 성공률이 $p$인 베르누이 시행을 $n$번 반복시행한 결과로 $X$는 $n$번 시행할 때의 성공횟수이다.
    $$f(x)=\binom{n}{x}p^x(1-p)^{n-x}, \ x = 1, \cdots, n$$

2.  포아송분포: $X\sim Poi(\lambda)$
    * 일정 기간 또는 특정 공간상에서 일어나는 독립적인 사건들의 횟수를 나타내는 분포이다.
    $$f(x)=\frac{\lambda^{x}}{x!}e^{-\lambda}, \ x=0,1,2,\cdots $$

3. 균일분포: $X \sim U(a,b)$
    * 확률변수 $X$가 $a$와 $b$사이에서 동일한 값을 가질 때 $X$를 균일분포를 따른다고 한다.
    $$f(x)=\frac{1}{b-a}, \ a<x<b$$

4. 베타분포: $X \sim Beta(\alpha,\beta)$
    * 정의역이 위, 아래로 모두 막혀있을 때 많이 사용되는 분포이다.
    * 밀도함수에 나오는 감마함수는 다음과 같은 특징이 있다.
    * $\Gamma(n) = (n-1)!$, 여기서 $n$은 자연수이다.
    * $\Gamma(1/2) = \sqrt{\pi}$
    $$f(x)= \frac{\Gamma(\alpha + \beta)}{\Gamma(\alpha) \Gamma(\beta)} x^{\alpha - 1} (1-x)^{\beta-1}, \ 0<x<1 $$

5. 지수분포: $X \sim \text{Exp}(\lambda)$
    * 하나의 사건이 일어난 후 독립인 그 다음 사건이 일어날 때까지 기다리는 시간을 모형화한 분포이다.
    $$f(x) = \lambda \exp(-\lambda x), \ x>0 \quad \text{또는} \quad f(x) = \dfrac{1}{\rho} \exp(-x/p), \ x>0$$

6. 정규분포: $X \sim N(\mu,\sigma^2)$
  * 평균과 표준편차에 의해 결정되는 분포로 종형의 확률밀도함수를 갖는다.
  $$f(x)=\frac{1}{\sqrt{2\pi\sigma^2}}\exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)$$


이제 SciPy 라이브러리를 이용하여 각 분포를 다루어보자.




## 2.2 이항분포

* 이항분포의 확률밀도 - `binom.pmf(k,n,p)`

  * `k`: 이항분포를 따르는 확률변수값  
  * `n`: 이항분포에서의 총 베르누이 시행횟수  
  * `p`: 베르누이 시행 시 성공확률  
  * $X\sim Bin(n,p)$ 일 때, $P(X=k)=\binom{n}{k}p^k (1-p)^{n-k}$의 값을 구한다.

* 이항분포의 누적확률 - `binom.cdf(k,n,p)`

  * `k`: 0부터 $k$까지의 누적확률을 구한다.  
  * $X \sim Bin(n,p)$일 때, $P(X\leq k)=\sum_{i=0}^{k}\binom{n}{i} p^i(1-p)^{n-i}$의 값을 구한다.

* 이항분포의 백분위수 - `binom.ppf(q,n,p)`

  * `q`: 구하고 싶은 분위수  
  * $X \sim Bin(n,p)$일 때, $P(X\leq x)=q$인 $x$의 값을 구한다.

* 이항분포의 난수생성 - `binom.rvs(n,p,size)`

  * `size`: 구하고 싶은 난수의 개수  
  * $X \sim Bin(n,p)$에서 `size`개의 난수를 생성한다.

  binom 모듈을 이용하는 방법은 아래의 코드를 참고한다.

```python
n = 10
p = 0.5
x = 5

# SciPy 라이브러리만 불러온 경우
import scipy
print(scipy.stats.binom.pmf(x, n, p))

# SciPy 라이브러리에서 stats 패키지를 불러온 경우
from scipy import stats
print(stats.binom.pmf(x, n, p))

# SciPy.stats 패키지에서 binom 모듈을 불러온 경우
from scipy.stats import binom
print(binom.pmf(x, n, p))
```

(이항분포 예제 1) $X \sim \text{Bin}(5, 1/10)$에서의 $P(X = 3)$의 값을 구하시오.

```python
import scipy
from scipy.stats import binom

# ------------------
# 코드 채우기
# ------------------
print(binom.pmf(k = 3, n = 5, p = 0.1))
print(round(binom.pmf(k = 3, n = 5, p = 0.1), ndigits = 3))

# round(수, 자릿수): 반올림함수
```

(이항분포 예제 2) $X \sim \text{Bin}(20, 1/10)$에서의 $P(X \le 3)$의 값을 구하시오.

```python
# ------------------
# 코드 채우기
# ------------------
print(binom.cdf(k = 3, n = 20, p = 0.1))
print(round(binom.cdf(k = 3, n = 20, p = 0.1), 3))
```

```python
# 파라미터 설정
n, p = 20, 0.1
k0 = 3

# 이항분포 support 및 pmf 설정
k = np.arange(0, n + 1)
pmf = binom.pmf(k, n=n, p=p)

# 시각화
plt.figure(figsize=(7, 4))

colors = ['salmon' if ki <= k0 else 'skyblue' for ki in k]

plt.bar(k, pmf, color=colors, alpha=0.5, edgecolor='k')

plt.xlim(-0.5, n + 0.5)
plt.ylim(0, pmf.max() * 1.25)
plt.xticks(k)
plt.yticks([])

plt.show()
```

## 2.3 포아송분포

* 포아송분포의 확률밀도 - `poisson.pmf(k, mu)`

  * `k`: 이항분포를 따르는 확률변수값  
  * `mu`: 포아송분포의 모수 $\lambda>0$  
  * $X\sim Poi(\lambda)$ 일 때, $P(X=x)=\frac{\lambda ^x}{x!}e^{-\lambda}$의 값을 구한다.

* 포아송분포의 누적확률 - `poisson.cdf(k ,mu)`

  * `k`: 0부터 $k$까지의 누적확률을 구한다.  
  * $X \sim Poi(\lambda)$일 때, $P(X\leq k)=\sum_{x=0}^{k}\frac{\lambda ^x}{x!}e^{-\lambda}$의 값을 구한다.

* 포아송분포의 백분위수 - `poisson.ppf(q,mu)`

  * `q`: 구하고 싶은 분위수  
  * $X \sim Poi(\lambda)$일 때, $P(X\leq x)=q$인 $x$의 값을 구한다.

* 포아송분포의 난수생성 - `poisson.rvs(mu,size)`

  * `size`: 구하고 싶은 난수의 개수  
  * $X \sim Poi(\lambda)$에서 `size`개의 난수를 생성한다.

(포아송분포 예제 2) $X \sim Poi(2)$에서 $P(X = 3)$의 값을 구하시오.

```python
from scipy.stats import poisson

# ------------------
# 코드 채우기
# ------------------
print(poisson.pmf(k = 3, mu = 2))

# round(수, 자릿수): 반올림함수
```

(포아송분포 예제 2) $X \sim Poi(2)$에서 $P(X \le 5)$의 값을 구하시오.

```python
# ------------------
# 코드 채우기
# ------------------

print(poisson.cdf(k = 5, mu = 2))
```

```python
# 파라미터 설정
lam = 2
k0 = 5

# 포아송 분포 support 및 pmf 설정
k = np.arange(0, 15)
pmf = poisson.pmf(k, mu=lam)

plt.figure(figsize=(7, 4))

colors = ['salmon' if ki <= k0 else 'skyblue' for ki in k]

plt.bar(k, pmf, color=colors, alpha=0.5, edgecolor='k')

plt.xlim(-0.5, k.max() + 0.5)
plt.ylim(0, pmf.max() * 1.25)
plt.xticks(k)
plt.yticks([])

plt.show()
```

(포아송분포 예제 3) Matplotlib를 이용하여 $X \sim Poi(\lambda)$일때, $\lambda=2,5,10$의 밀도함수를 그리시오.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 20, 1)  # 0부터 20까지 1간격으로 배열 생성 np.arange(21)
y1 = poisson.pmf(x, 2)  # lambda=2 일때 밀도함수 값
y2 = poisson.pmf(x, 5)
y3 = poisson.pmf(x, 10)

plt.bar(x, y1, color='salmon', alpha=0.3, label='lambda=2')
plt.bar(x, y2, color='purple', alpha=0.3, label='lambda=5')
plt.bar(x, y3, color='skyblue', alpha=0.3, label='lambda=10')
plt.legend()
plt.show()
```

## 2.4 균일분포

`from scipy.stats import uniform`을 통해 균일분포 함수들을 사용할 수 있다.  
python에서는 location 변수와 scale 변수를 통해 Uniform(loc, loc+scale)의 형태로 균일분포를 표현한다.

$$X\sim U(a,b)$$

* 균일분포의 확률밀도 - `uniform.pdf(x,loc,scale)`

  * `x`: 균일분포를 따르는 확률변수값  
  * `loc`: 균일분포에서 하한값$(a)$  
  * `sclae`: 균일분포에서 상한값$-$하한값 $(b-a)$

* 균일분포의 누적확률 - `uniform.cdf(x,loc,scale)`

  * `x`: 여기서의 `x` 값은 확률변수값 `x` 까지의 누적확률을 구한다는 의미이다.

* 균일분포의 백분위수 - `uniform.ppf(q,loc,scale)`

  * `q`: 구하고 싶은 분위수

* 균일분포의 난수생성 - `uniform.rvs(loc,scale,size)`

  * $X \sim U(a,b)$에서 `size`개의 난수를 생성한다.

(균일분포 예제 1) $X \sim U(0,1)$에서 $P(X \le 0.3)$인 $x$의 값과 무작위로 10개의 표본을 추출한 결과를 나타내시오.

```python
from scipy.stats import uniform

# ------------------
# 코드 채우기
# ------------------

print(uniform.ppf(q = 0.3, loc = 0, scale = 1))
print(uniform.rvs(loc = 0 , scale = 1, size = 10))
```

## 2.5 베타분포

`from scipy.stats import beta`을 통해 베타분포 함수들을 사용할 수 있다.

* 베타분포의 확률밀도 - `beta.pdf(x,a,b)`

  * `x`: 베타분포를 따르는 확률변수값  
  * `a`: 첫번째 모수($\alpha$), `b`: 두번째 모수($\beta$)

* 베타분포의 누적확률 - `beta.cdf(x,a,b)`

  * $X \sim Beta(\alpha,\beta)$의 확률밀도함수 $P(X \leq x)$의 값을 구한다.

* 베타분포의 백분위수 - `beta.ppf(q,a,b)`

  * $X \sim Beta(\alpha,\beta)$일 때, $P(X \leq x)=q$를 만족하는 x의 값을 구한다.

* 베타분포의 난수생성 - `beta.rvs(a,b,size)`

  * $X \sim Beta(\alpha,\beta)$에서 `size`개의 난수를 생성한다.

(베타분포 예제 1) $X \sim Beta(1,10)$에서, $P(X \le 0.1)$값을 구하시오.

```python
from scipy.stats import beta

# ------------------
# 코드 채우기
# ------------------
print(beta.cdf(x = 0.1, a = 1, b = 10))
```

```python
x0 = 0.1

# 베타분포 support 및 pdf 설정
x = np.linspace(0, 1, 500)
pdf_11 = beta.pdf(x, a=1, b=10)

# 시각화
plt.figure(figsize=(7, 4))

plt.plot(x, pdf_11, color='k', lw=2)
plt.fill_between(x, 0, pdf_11, where=(x <= x0), color='salmon', alpha=0.5)
plt.fill_between(x, 0, pdf_11, where=(x > x0), color='skyblue', alpha=0.5)

plt.xlim(0, 1)
plt.ylim(0, pdf_11.max() * 1.25)

plt.show()
```

(베타분포 예제 2) Matplotlib 패키지를 이용한 베타분포, $Beta(\alpha,\beta), ~ \alpha = 1,4, \beta = 1,3,5$의 밀도함수를 그리시오.

```python
import numpy as np
import matplotlib.pyplot as plt

a = [1, 4]
b = [1, 3, 5]
x = np.arange(0, 1, 0.01)

# 반복문을 이용해 간단하게 코드 작성 가능
for i in a:
    for j in b:
        # ------------------
        # 코드 채우기
        # ------------------
        y = beta.pdf(x = x, a = i, b = j)
        plt.plot(x, y, label='a={}, b={}'.format(i, j))
plt.legend()
plt.show()
```

(참고) $X \sim Beta(1, 1)$의 밀도함수는 $f(x) = 1, 0<x<1$이므로 균등분포 $X \sim U(0,1)$와 같다.

## 2.6 지수분포

`from scipy.stats import expon`을 통해 지수분포 함수들을 사용할 수 있다.  

Python의 지수분포는 `scale` 파라미터를 사용하며,  
$X \sim Exp(\lambda)$일 때 확률밀도함수
$f(x)=\lambda e^{-\lambda x},\; x>0$는 $\rho = 1/\lambda$를 사용하면 $f(x)=\frac{1}{\rho}e^{-x/\rho},\; x>0$로도 표현된다.  
이때 Python에서는 $\text{scale} = \rho = 1/\lambda$에 해당한다.

* 지수분포의 확률밀도 - `expon.pdf(x,loc,scale)`

  * `x`: 지수분포를 따르는 확률변수값  
  * `loc`: 지수분포의 시작점 (보통 0)  
  * `scale`: 지수분포의 모수 $(\rho = 1/\lambda)$  

* 지수분포의 누적확률 - `expon.cdf(x,loc,scale)`

  * $X \sim Exp(\rho)$ (scale 파라미터 $\rho$)일 때,  
    $P(X \leq x)=1-e^{-x/\rho}$의 값을 구한다.

* 지수분포의 백분위수 - `expon.ppf(q,loc,scale)`

  * `q`: 구하고 싶은 분위수  
  * $X \sim Exp(\rho)$일 때,  
    $P(X \leq x)=q$를 만족하는 $x$의 값을 구한다.

* 지수분포의 난수생성 - `expon.rvs(loc,scale,size)`

  * $X \sim Exp(\rho)$에서 `size`개의 난수를 생성한다.

(지수분포 예제 1) 확률변수 $X$는 scale 파라미터가 $\rho=4$인 지수분포를 따른다고 하자. $P(X \le 2)$와 $P(X \ge 6)$를 구하시오.

```python
from scipy.stats import expon

# P(X <= 2)
# ------------------
# 코드 채우기
# ------------------
p1 = expon.cdf(x = 2, scale = 4)

# P(X >= 6)
# ------------------
# 코드 채우기
# ------------------

p2 = 1 - expon.cdf(x = 6, scale = 4)

print(p1)
print(p2)
```

```python
# 파라미터 설정
scale = 4

# 지수분포 support 및 pdf 설정
x = np.linspace(0, 20, 500)
pdf = expon.pdf(x, scale=scale)

x_left = 2
x_right = 6

# 시각화
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

# P(X <= 2)
ax[0].plot(x, pdf, color='k', lw=2)
ax[0].fill_between(x, 0, pdf, where=(x <= x_left), color='salmon', alpha=0.5)
ax[0].fill_between(x, 0, pdf, where=(x > x_left), color='skyblue', alpha=0.5)

ax[0].set_xlim(0, 20)
ax[0].set_ylim(0, pdf.max() * 1.25)

# P(X >= 6)
ax[1].plot(x, pdf, color='k', lw=2)
ax[1].fill_between(x, 0, pdf, where=(x >= x_right), color='salmon', alpha=0.5)
ax[1].fill_between(x, 0, pdf, where=(x < x_right), color='skyblue', alpha=0.5)

ax[1].set_xlim(0, 20)
ax[1].set_ylim(0, pdf.max() * 1.25)

plt.tight_layout()
plt.show()
```

(지수분포 예제 2) Matplotlib 패키지를 이용하여 지수분포 $Exp(\rho)$의 확률밀도함수를 그려보자. scale 파라미터가 $\rho = 0.5, 1, 2$인 경우에 대해, $x \in [0, 5]$ 구간에서 각 지수분포의 확률밀도함수를 한 그래프에 함께 나타내어라.

```python
scale = [0.5, 1, 2]
x = np.arange(0, 5, 0.01)

# 반복문을 이용해 간단하게 작성
for s in scale:
    # ------------------
    # 코드 채우기
    # ------------------
    y = expon.pdf(x = x, scale = s)
    plt.plot(x, y, label='scale={}'.format(s))

plt.legend()
plt.show()
```

## 2.7 정규분포

`from scipy.stats import norm`을 통해 정규분포 함수들을 사용할 수 있다.

* 정규분포의 확률밀도 - `norm.pdf(x,loc,scale)`

  * `x`: 정규분포를 따르는 확률변수값  
  * `loc`: 평균($\mu$), `scale`: 표준편차($\sigma$)

* 정규분포의 누적확률 - `norm.cdf(x,loc,scale)`

  * $X \sim N(\mu,\sigma^2)$의 확률밀도함수 $P(X \leq x)$의 값을 구한다.

* 정규분포의 백분위수 - `norm.ppf(q,loc,scale)`

  * $X \sim N(\mu,\sigma^2)$일 때, $P(X \leq x)=q$를 만족하는 x의 값을 구한다.

* 정규분포의 난수생성 - `norm.rvs(loc,scale,size)`

  * $X \sim N(\mu,\sigma^2)$에서 `size`개의 난수를 생성한다.

(정규분포 예제 1) $X \sim N(200, 2^2)$에서 $P(X \ge 201)$ 값을 구하시오.

```python
from scipy.stats import norm

# ------------------
# 코드 채우기
# ------------------
print(1 - norm.cdf(x = 201, loc = 200, scale = 2))
```

```python
# 파라미터 설정
mu, sigma = 200, 2
x0 = 201

# 정규분포 구간 및 pdf 설정
x = np.linspace(190, 210, 500)
pdf = norm.pdf(x, loc=mu, scale=sigma)

# 시각화
plt.figure(figsize=(7, 4))

plt.plot(x, pdf, color='k', lw=2)
plt.fill_between(x, 0, pdf, where=(x >= x0), color='salmon', alpha=0.5)
plt.fill_between(x, 0, pdf, where=(x < x0), color='skyblue', alpha=0.5)

plt.xlim(190, 210)
plt.ylim(0, pdf.max() * 1.25)

plt.show()
```

(참고) $X \sim N(200, 2^2)$에서의 $P(X \ge 201) = 1 - P(X \le 201) = 1 - P(Z \le (201 - 200)/2) = 1 - P(Z \le -1/2)$ 값이다.

(정규분포 예제 2) Matplotlib 패키지를 이용하여 표준정규분포 $N(0,\sigma^2)$의
확률밀도함수를 그려보자. 분산이 $\sigma^2 = 0.5, 1, 2$인 경우에 대해, $x \in [-5, 5]$ 구간에서 각 정규분포의 확률밀도함수를 한 그래프에 함께 나타내어라.

```python
variance = [0.5, 1, 2]
x = np.arange(-5, 5, 0.01)

# 반복문을 이용해 간단하게 작성
for v in variance:
    # ------------------
    # 코드 채우기
    # ------------------
    y = norm.pdf(x = x, loc = 0, scale = np.sqrt(v))
    plt.plot(x, y, label='variance={}'.format(v))

# x = 0 축 추가
plt.axvline(x=0, linestyle='--', alpha=0.7)

plt.legend()
plt.show()
```
